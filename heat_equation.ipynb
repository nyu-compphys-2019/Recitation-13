```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 0.01 
D = 4.25e-6
N = 100
a = L/N #gridspacing
h = 1e-4 #timestep
eps = h/1000 
```

```python
Tlo = 0.0
Tmid = 20.0
Thi = 50.0 
```

```python
t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0 
t5 = 10.0
tend = t5 + eps
```

```python
T = np.empty(N+1, float)
T[0] = Thi
T[N] = Tlo
T[1:N] = Tmid
Tp = np.empty(N+1, float)
Tp[0] = Thi
Tp[N] = Tlo 
```

```python
t = 0.0
c = h*D /(a**2)
c
```

```python
while t<tend: 
    for i in range(1, N):
        Tp[i] = T[i] + c*(T[i+1] + T[i-1] - 2*T[i])
    
    T = Tp
    t += h
    
    if np.abs(t-t1) < eps:
        plt.plot(T)
    if np.abs(t-t2) < eps:
        plt.plot(T)
    if np.abs(t-t3) < eps:
        plt.plot(T)
    if np.abs(t-t4) < eps:
        plt.plot(T)
    if np.abs(t-t5) < eps:
        plt.plot(T)
```
